```python
from dataclasses import dataclass


class MyClassicNumber:
    def __init__(self, value: int):
        self.value = value

    def __eq__(self, other: 'MyClassicNumber'):
        return self.value == other.value

    def __lt__(self, other: 'MyClassicNumber'):
        return self.value < other.value

    def __gt__(self, other: 'MyClassicNumber'):
        return self.value > other.value

    def __le__(self, other: 'MyClassicNumber'):
        return self.value <= other.value

    def __ge__(self, other: 'MyClassicNumber'):
        return self.value >= other.value

    def __ne__(self, other: 'MyClassicNumber'):
        return self.value != other.value

    def __repr__(self) -> str:
        return f'{self.value}'


@dataclass(order=True)
class MyDataClassNumber:
    value: int


if __name__ == "__main__":
    num1 = MyClassicNumber(17)
    num2 = MyClassicNumber(17)
    print(num1 < num2) # False
    print(num1 > num2) # False
    print(num1 <= num2) # True
    print(num1 >= num2) # True
    print(num1 == num2) # True
    print(num1 != num2) # False

    num1 = MyDataClassNumber(10)
    num2 = MyDataClassNumber(11)
    print(num1 < num2) # True
    print(num1 > num2) # False
    print(num1 <= num2) # True
    print(num1 >= num2) # False
    print(num1 == num2) # False
    print(num1 != num2) # True
```

```python
from dataclasses import dataclass


class MyClassicNumber:
    def __init__(self, w: int, h:int):
        self.w = w
        self.h = h

    def square(self) -> int:
        return self.w * self.h

    def __eq__(self, other: 'MyClassicNumber'):
        return self.square() == other.square()

    def __lt__(self, other: 'MyClassicNumber'):
        return self.square() < other.square()

    def __gt__(self, other: 'MyClassicNumber'):
        return self.square() > other.square()

    def __le__(self, other: 'MyClassicNumber'):
        return self.square() <= other.square()

    def __ge__(self, other: 'MyClassicNumber'):
        return self.square() >= other.square()

    def __ne__(self, other: 'MyClassicNumber'):
        return self.square() != other.square()

    def __repr__(self) -> str:
        return f'Rectangle(w:{self.w}, h: {self.h}' \
               f'square {self.square()})'


@dataclass(order=True)
class MyDataClassRectangle:
    h: int
    w: int

    def square(self) -> int:
        return self.w * self.h


if __name__ == "__main__":
    rec1 = MyClassicNumber(3, 5)
    rec2 = MyClassicNumber(5, 3)
    print(rec1 < rec2) # False
    print(rec1 > rec2) # False
    print(rec1 <= rec2) # True
    print(rec1 >= rec2) # True
    print(rec1 == rec2) # True
    print(rec1 != rec2) # False

    print()
    rec1 = MyDataClassRectangle(3, 5)
    rec2 = MyDataClassRectangle(5, 3)
    print(rec1 < rec2) # True
    print(rec1 > rec2) # False
    print(rec1 <= rec2) # True
    print(rec1 >= rec2) # False
    print(rec1 == rec2) # False
    print(rec1 != rec2) # True

    rec1 = MyClassicNumber(3, 5)
    print(rec1.square() == rec2.square()) 
```

```python
from dataclasses import dataclass


@dataclass
class Employee:
    name: str
    employee_id: int
    age: int

    def get_age(self) -> int:
        return self.age

    def get_salary(self) -> int:
        return (self.employee_id + self.age) * 1000


@dataclass
class Driver(Employee):
    auto_type: str
    salary_coeff: float = 0.8 # по умолчанию

    def get_salary(self) -> int:
        return int(super().get_salary() * self.salary_coeff)


@dataclass
class Waiter(Employee):
    ...


if __name__ == '__main__':
    employee_list: list[Employee] = [
        Driver('Alex', 12, 45, 'Lada'),
        Waiter('Max', 5, 19),
        Waiter('Nikita', 6, 20),
        Driver('Gleb', 11, 34, 'Oka', 0.76),
    ]

    employee_type = ''
    for it in employee_list:
        if isinstance(it, Driver):
            employee_type = 'Driver'
        if isinstance(it, Waiter):
            employee_type = 'Waiter'
        print(f'{employee_type} {it.name} have salary '
              f'= {it.get_salary()}')
```

```python
from dataclasses import dataclass


@dataclass
class Employee:
    name: str
    employee_id: int
    age: int

    def __post_init__(self):
        self.salary = (self.employee_id + self.age) * 1000

    def get_age(self) -> int:
        return self.age

    def get_salary(self) -> int:
        return self.salary


@dataclass
class Waiter(Employee):

    def __post_init__(self):
        super().__post_init__()


if __name__ == '__main__':
    waiter = Waiter('Nikita', 6, 20)
    print(f'Waiter {waiter.name} have salary '
          f'= {waiter.get_salary()}')
```

```python
from dataclasses import dataclass, field


@dataclass
class Employee:
    name: str = 'Alex'
    employee_id: int = 6
    age: int = field(repr=False, default=14)


if __name__ == '__main__':
    print(Employee())
    print(Employee('Max', 10))
# Employee(name='Alex', employee_id=6)
# Employee(name='Max', employee_id=10)
```

```python
from dataclasses import dataclass, field


@dataclass(order=True)
class Employee:
    name: str = field(compare=False)
    employee_id: int = 6
    age: int = field(compare=False, repr=False, default=14)


if __name__ == '__main__':
    employee1 = Employee('Alex')
    employee2 = Employee('Max', 10)
    
    print(employee1) # Employee(name='Alex', employee_id=6)
    print(employee2) # Employee(name='Max', employee_id=10)
    
    
    print(employee1 == employee2) # False
    print(employee1 <= employee2) # True
```
